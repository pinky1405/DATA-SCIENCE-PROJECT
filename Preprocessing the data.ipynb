```python
# create a list pf feature to dummy
todummy_list = ['City','Gender','College name','University Name','Degree','Major/Area of Study','Course Type','Which-year are you studying in?','Areas of interest','Current Employment Status','Programming Language Known other than Java (one major)','How Did You Hear About This Internship?']
```

```python
#fuction to dummy all
def dummy_df(df, todummy_list):
    for x in todummy_list:
        dummies= pd.get_dummies(df[x], prefix=x ,dummy_na=False)
        df=df.drop(x,1)
        df=pd.concat([df,dummies], axis=1)
    return df 
```

# Implementing Logistic Regression

```python
# creating a dependent and independent variable
x = file.drop(['Label'], axis=1)
y = file['Label']
x.shape, y.shape
```

```python
# Importing the train test split function
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(x,y, random_state = 56)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
cols = train_x.columns
cols
```

```python
train_x_scaled = scaler.fit_transform(train_x)
train_x_scaled = pd.DataFrame(train_x_scaled, columns=cols)
train_x_scaled.head()
```

```python
test_x_scaled = scaler.transform(test_x)
test_x_scaled = pd.DataFrame(test_x_scaled, columns=cols)
test_x_scaled.head()
```

```python
#importing Logistic Regression and metric F1-score
from sklearn.linear_model import LogisticRegression as LogReg
from sklearn.metrics import f1_score
```

```python
# Creating instance of Logistic Regresssion
logreg = LogReg()

# Fitting the model
logreg.fit(train_x, train_y)
```

```python
# Predicting over the Train
train_predict = logreg.predict(train_x)
train_predict
```

```python
# Calculating f1-score
k = f1_score(train_predict, train_y)
print('Training f1_score', k )
```

```python
# Predicting over the Test Set and f1-score
test_predict = logreg.predict(test_x)
k = f1_score(test_predict, test_y)
print('Test f1_score ', k )
```

# Making predictions using predict_proba function

```python
# Predicting over the Train
train_predict = logreg.predict_proba(train_x)
train_predict
```

```python
train_preds = train_predict[:,1]
train_preds
```

```python
for i in range(0, len(train_preds)):
  if(train_preds[i]>0.55):
    train_preds[i] = 1
  else:
    train_preds[i] = 0
```

```python
# Calculating f1-score
k = f1_score(train_preds, train_y)
print('Training f1_score', k )
```

# Confusion matrix

```python
from sklearn.metrics import confusion_matrix
cf= confusion_matrix(test_y, test_predict)
print(cf)
```

```python
from sklearn.metrics import classification_report as rep
print(rep( test_y , test_predict ))
```

# KNN model Implementation


```python
#seperating independent and dependent variables
x = file.drop(['Label'], axis=1)
y = file['Label']
x.shape, y.shape
```

```python
## Importing the MinMax Scaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x = pd.DataFrame(x_scaled, columns = x.columns)
```

```python
x.head()
```

```python
# Importing the train test split function
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(x,y, random_state = 56, stratify=y)
```

```python
#importing KNN classifier and metric F1score
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.metrics import f1_score
```

```python
# Creating instance of KNN
clf = KNN(n_neighbors = 10)

# Fitting the model
clf.fit(train_x, train_y)

# Predicting over the Train Set and calculating F1
test_predict = clf.predict(test_x)
k = f1_score(test_predict, test_y)
print('Test F1 Score    ', k )
```

```python
#seperating independent and dependent variables
x = file.drop(['Label'], axis=1)
y = file['Label']
x.shape, y.shape
```

```python
## Importing the MinMax Scaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x = pd.DataFrame(x_scaled, columns = x.columns)
```

```python
# Importing the train test split function
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(x,y, random_state = 56)
```

```python
from sklearn.svm import SVC
svclassifier = SVC(kernel='linear')
svclassifier.fit(train_x,train_y)
```

```python
y_pred = svclassifier.predict(test_x)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import f1_score
print(confusion_matrix(test_y,y_pred))
print(classification_report(test_y,y_pred))
```

```python
# Calculating f1-score
k = f1_score(y_pred, test_y)
print('Training f1_score', k )
```
